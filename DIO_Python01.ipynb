```python
def exibir_mensagem():
  print("Seja bem vindo!")

exibir_mensagem()

def exibir_mensagem_2(nome):
  print(f"Seja bem vindo {nome}")

exibir_mensagem_2("João")

def exibir_mensagem_3(nome2="Anônimo"):
  print(f"Seja bem vindo, {nome2}")

exibir_mensagem_3()
exibir_mensagem_3(nome2="João P.")
```

```python
def somar(a, b):
  return a+b

def multiplicar(a, b):
  return a*b

def exibir_resultado(a, b, funcao):
  resultado = funcao(a, b)
  print(f"O resultado da operação é {resultado}")

exibir_resultado(5, 10, somar)
exibir_resultado(10, 20, multiplicar)
```

```python
salario = 2000

def salario_bonus(bonus):
  global salario
  salario += bonus
  return salario

salario_bonus(500)
```
